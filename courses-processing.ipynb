```python
import pandas as pd

df = pd.read_csv('coursesW23_raw.csv')
```

```python
# Remove rows with null or empty values in the CRN column
df = df.dropna(subset=['CRN'])
df = df[df['CRN'].astype(str).str.strip() != '']
# Reset the index of the DataFrame
df = df.reset_index(drop=True)
df.head()
```

```python
my_list = df.columns.values.tolist()
print(my_list)
```

```python
df = df.drop(['WL Cap', 'WL Act', 'WL Rem', 'XL Cap', 'XL Rem', 'XL Cap', 'Cmp', 'Rem', 'Act', 'XL Act'], axis=1)
```

```python
# index=False to prevent saving index vals as column
df.to_csv('w23courses_filtered.csv', index=False)
df.head(n=5)
```
