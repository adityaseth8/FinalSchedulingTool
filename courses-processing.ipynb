```python
import pandas as pd

df = pd.read_csv('coursesW23_raw.csv')
```

```python
# Remove rows with null or empty values in the CRN column
df = df.dropna(subset=['CRN'])
df = df[df['CRN'].astype(str).str.strip() != '']
# Reset the index of the DataFrame
df = df.reset_index(drop=True)
df.head()
```

```python
my_list = df.columns.values.tolist()
print(my_list)
```

```python
df = df.drop(['Select', 'WL Cap', 'WL Act', 'WL Rem', 'XL Cap', 'XL Rem', 'XL Cap', 'Cmp', 'Rem', 'Act', 'XL Act'], axis=1)
df.head()
```

```python
grouped = df.groupby(['Subj', 'Crse', df['Sec'].str[0]])['Cap'].sum()
grouped.head()
```

```python
df['Cap'] = df['Cap'].astype(int)
```

```python
# combine the same sections

# Group the DataFrame by 'Subj', 'Crse', and the first letter of 'Sec'
grouped = df.groupby(['Subj', 'Crse', df['Sec'].str[0]])

# Sum the 'Cap' values within each group
df_combined = grouped['Cap'].sum().reset_index()

df_combined.head()
```

```python
merged_df = pd.merge(df_combined, df, on=['Subj', 'Crse'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.drop(['Cred', 'Cap_y', 'Sec_y'], axis=1)
merged_df.head()
```

```python
merged_df.rename(columns={"Sec_x": "Sec", "Cap_x": "Cap"})
```

```python
# index=False to prevent saving index vals as column
merged_df.to_csv('w23courses_filtered.csv', index=False)
merged_df.head(n=5)
```
